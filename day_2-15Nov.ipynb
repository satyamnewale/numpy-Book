difference between dot and @
1. dot is applied to vector-vector, vector-matrix, matrix-matrix.
2. @ can also be doen with all.
3. dot choose behavior depending on shape (special case)
4. @ follows (mXn)@(nXp) and gives (mXp)
5. in dot vector-vector gives scalar
6. edge case :- in (1X3)(3X1) matrix output is matrix of(1X1) but numpy    convert it to scalar value and consider it as dot product.

---
1D array in matmul

---


1. (n,) @ (n,)	:      allowed	     -     (1×n) @ (n×1) converted to row and col vector




2. (n,) @ (n,k):	   NOT allowed    -  	no valid way




3. (n,m) @ (m,):	   NOT allowed	    -  no valid way



4. (n,) @ (n,1)	:   allowed	   -       (1×n) @ (n×1)  converted to row vector




5. (n,1) @ (n,)	:   NOT allowed	   -    becomes (n×1) @ (nx1) converted to col vector → outer product not satisdy  (mXn)@(nXp)





6. (n,) @ (1,n)	:   NOT allowed	   -    becomes (1xn) @ (1xn) converted to row vector → outer product not satisfy  (mXn)@(nXp)


---
1. rule: if value is at left side the (n,) -> (1,n)
2. rule: if value is at right side the (n,) -> (n,1)


---
element wise multiplication vs (dot and @)
1. in element wise only element and scalar value are multiplied
2. in dot and @ elements are multiplied and then added.
---

cosine similarity and angle:

1. +1 → perfectly same direction
2. 0 → 90° apart → perpendicular → no similarity
3. –1 → opposite direction

angle:

1. 0° → identical direction
2. Small angle (<30°) → very similar
3. 90° → unrelated
4. greater than 120° → mostly opposite
5. 180° → exactly opposite direction

---

multiplication rules:
1. always check inner dimensions ,they should be same (mxn)@(nxp).
2. iuter shape should be (mxp).
3. Row of A interact with column of B.
4. not commutative : (mxn) != (nxm).
5. identity rule: I x A(nxm) = A.
---

broadcasting:
1. compare shape from right to left, last dim are equal. eg: (5,3) and (4,3) accepted.
2. they must be equal or one must be 1. eg:- (4,1) and (4,3) accepted, because 1 stretchs to 4.
3, missing dimensions is treaded as 1. eg:- (3,) and (2,3) accepted, (3,) treated as (1,3) row vector.
eg: (3,) + (2,) cannot broadcast.
eg: (2,3) - (3,) not accepted as convert to (3,1) due to left side as it becomes col vector and dont follow multiplication rule.

```
import numpy as np

arr = np.array([[1,2],[2,4],[4,8]])
print(f"array : {arr}")

#vector maths:
#element-wise:-
arr_add = arr + 5 # expected [[6,7],[7,9],[9,13]]
print(f"addition :{arr_add}")

arr_sub = arr - 3
print(f"subtraction : {arr_sub}")

arr_mul = arr * 2
print(f"multiplication : {arr_mul}")

arr_div = arr / 2 #scalar multiplication element wise
print(f"division : {arr_div}")

arr_vect1 = np.array([5,2,1])
arr_vect2 = np.array([1,2,3])

print(f"addition of vectors : {arr_vect1+arr_vect2}")
print(f"subtraction of vectors : {arr_vect1-arr_vect2}")
print(f"division of vectors : {(arr_vect1/arr_vect2).astype(np.int32)}")
print(f"multiplication of vectors : {arr_vect1*arr_vect2}")

#dot product and @
arr1 = np.array([[1,2,3]])
arr2 = np.array([[1],[2],[3]])
arr_vect = np.array([1,2,3])

print(f"shape of arr1:{arr1.shape}, dim:{arr1.ndim}")
print(f"shape of arr2:{arr2.shape}, dim:{arr2.ndim}")
print(f"shape of arr_vect:{arr_vect.shape}, dim:{arr_vect.ndim}")

#dot product
print(f"dot product : {np.dot(arr1,arr2)}, dim {np.dot(arr1,arr2).ndim}, shape:{np.dot(arr1,arr2).shape}")
print(f"@ product : {arr2@arr1}, dim {(arr2@arr1).ndim}, shape:{(arr2@arr1).shape}")
print(f"dot product of vector and matrix: {np.dot(arr_vect,arr2)}, dim {np.dot(arr_vect,arr2).ndim}, shape:{np.dot(arr_vect,arr2).shape}, gives scalar value.")
print(f"@ product of vector and matrix: {arr_vect@arr2},dim {(arr_vect@arr2).ndim}, shape:{(arr_vect@arr2).shape}, A 1D vector behaves like a row vector during @ (matmul) as it is at left side .NumPy automatically upgrades (3,) into (1,3) for matrix multiplication.")
print(f"@ product of vector and matrix: {arr1@arr_vect},dim {(arr1@arr_vect).ndim}, shape:{(arr1@arr_vect).shape}, A 1D vector behaves like a rcol vector during @ (matmul) as it is at right side .NumPy automatically upgrades (3,) into (3,1) for matrix multiplication.")

try:
  print(f"@ product of vector and matrix: {arr_vect@arr1},dim {(arr_vect@arr1).ndim}, shape:{(arr_vect@arr1).shape}")
except:
  print("@ product of vector and matrix: {arr_vect@arr1} ,A 1D vector behaves like a row vector during @ (matmul) as it is at left side .NumPy automatically upgrades (3,) into (1,3) for matrix multiplication but the dimensions dont satisfy (mXn)@(nXp) rule.")

try:
  print(f"@ product of vector and matrix: {arr2@arr_vect},dim {(arr2@arr_vect).ndim}, shape:{(arr2@arr_vect).shape}")
except:
  print("@ product of vector and matrix: {arr2@arr_vect} ,A 1D vector behaves like a col vector during @ (matmul) as it is at right side .NumPy automatically upgrades (3,) into (3,1) for matrix multiplication but the dimensions dont satisfy (mXn)@(nXp) rule.")

print(f"dot product : {np.dot(arr1,arr_vect)}, dim {np.dot(arr1,arr_vect).ndim}, shape:{np.dot(arr1,arr_vect).shape}")

try:
  print(f"dot product : {np.dot(arr2,arr_vect)}, dim {np.dot(arr2,arr_vect).ndim}, shape:{np.dot(arr2,arr_vect).shape}")
except:
  print("dot product : {np.dot(arr2,arr_vect)} dont satisfy rule ((mxn) dot (nxp))")
print("-----------------------------------------------------------------------------")

#slicing col and row wise
print("slicing :")

arr2D = np.array([[1,2,3,4],[5,6,7,8],[2,4,6,8],[1,3,5,7]])

print(f"row slicing : {arr2D[0:2,:]}")
print(f"col slicing : {arr2D[:,1:3]}")
print(f"row and col slicing : {arr2D[1:3,2:4]}")
print("-----------------------------------------------------------------------------")

print("transpose :")
array1 = np.array([[1,2,3],[1,4,7]])
array2 = np.array([[1,2,3]])
array3 = np.array([[1,2,3],[1,4,7],[3,4,5],[4,5,8]])
print(f"array1 : {array1}, shape of array1: {array1.shape}, array2: {array2}, shape of array2: {array2.shape}, array3: {array3}, shape of array3: {array3.shape}")
print(f"transpose of array1: {array1.T}, shape of T array1: {(array1.T).shape}, T of array2: {array2.T}, shape of T array2: {(array2.T).shape}, T of array3: {array3.T}, shape of T array3: {(array3.T).shape}")
print("transpose means to switch its rows to columns and columns to rows.")
print("-----------------------------------------------------------------------------")

print("norms , distance between vectors, cosine similarity and angle")
# there are three types of norms (Euclidean is mostly used)
# L2(euclidean norm = √(x₁² + x₂² + x₃² + …))
# L1(|x₁| + |x₂| + |x₃| + …)
# L∞(max(|x₁|, |x₂|, |x₃|, …))

arr_1 = np.array([3, 2,-4])
arr_2 = np.array([4, 0,-1])

#L2 manually calculate
L2_manual = (3**2 + 2**2 + (-4)**2)**0.5
print(f"L2 manually calculate : {L2_manual}, this is distance of vector from origin.")

#L2 with method
L2_linalg = np.linalg.norm(arr_1)
print(f"L2 with method : {L2_linalg}")

L1_manual = abs(3) + abs(2) + abs(-4)
print(f"L1 manually calculate : {L1_manual}")

L1_linalg = np.linalg.norm(arr_1,1)
print(f"L1 with method : {L1_linalg}")

L3_manual = max(abs(3),abs(2),abs(-4))
print(f"L3 manually calculate : {L3_manual}")

L3_linalg = np.linalg.norm(arr_1,np.inf)
print(f"L3 with method : {L3_linalg}")

# distance between vector
#formula ||v1-v2|| manually and with numpy method
dist_manual = ((3-4)**2 + (2-0)**2 + (-4-(-1))**2)**0.5
print(f"distance between vector manually : {dist_manual}")

dist_linalg = np.linalg.norm(arr_1-arr_2)
print(f"distance between vector with method : {dist_linalg}")

#cosine similarity manually and with methods
cosine_manual = (3*4 + 2*0 + (-4)*(-1))/((3**2 + 2**2 + (-4)**2)**0.5 * (4**2 + 0 +(-1)**2)**0.5)
print(f"cosine similarity manually : {cosine_manual}")

cosine_linalg = np.dot(arr_1,arr_2)/(np.linalg.norm(arr_1)*np.linalg.norm(arr_2))
print(f"cosine similarity with method : {cosine_linalg}")
print("as cosine similarity is 0.72 which is near 1 means the vectors are in acute angle")

angle = np.degrees(np.arccos(cosine_linalg))
print(f"angle between vectors : {angle}, the angle is 44 degree so it very similar.")

#broadcasting:
```
